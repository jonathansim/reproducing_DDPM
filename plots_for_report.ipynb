```python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import inset_axes


# Data
epochs = [1, 25, 50, 75]
fid_scores = {
    "linear_heads_8": [165.63026288120784, 28.578087424705398, 18.981849308074096, 12.153196400320432],
    "cosine_heads_8": [193.2679935346893, 20.9626612834261, 12.784471419321257, 14.496150726788745],
    "linear_heads_4": [170.04469736268175, 30.281336636391686, 19.34383023149735, 11.262894094318604],
    "cosine_heads_4": [190.52100654833438, 20.365324135599053, 16.374350071521288, 11.72226122140874],
    "linear_heads_1": [155.02214900584394, 29.16643743896486, 18.565640542029456, 11.224682753933491],
    "cosine_heads_1": [182.90545942044724, 22.151145376909163, 18.349605153603676, 13.875475483862495],
}

# Define a color palette for each group
#linear_colors = ["deepskyblue", "blue", "darkturquoise"]  # Distinct colors for linear
# cosine_colors = ["red", "darkred", "crimson"]  # Distinct colors for cosine

linear_colors = ["blue", "red", "green"]  # Distinct colors for linear
cosine_colors = ["blue", "red", "green"]  # Distinct colors for cosine


# Define line styles for distinction
line_styles = {
    "linear": "-",
    "cosine": "--",
}

# Create main plot
fig, ax = plt.subplots(figsize=(10, 6))
for i, (model, scores) in enumerate(fid_scores.items()):
    if "linear" in model:
        color = linear_colors[i % len(linear_colors)]
        linestyle = line_styles["linear"]
    elif "cosine" in model:
        color = cosine_colors[i % len(cosine_colors)]
        linestyle = line_styles["cosine"]
    else:
        color = "gray"
        linestyle = "-"
    ax.plot(epochs, scores, marker="o", label=model, color=color, linestyle=linestyle)

ax.set_title("FID Scores Over Epochs for Different Models")
ax.set_xlabel("Epoch")
ax.set_ylabel("FID Score")
ax.set_xticks(epochs)
ax.legend(title="Models")
ax.grid()

# Create inset zoomed plot
inset_ax = inset_axes(ax, width="40%", height="40%", loc="center right")  # Inset size and position
for i, (model, scores) in enumerate(fid_scores.items()):
    if "linear" in model:
        color = linear_colors[i % len(linear_colors)]
        linestyle = line_styles["linear"]
    elif "cosine" in model:
        color = cosine_colors[i % len(cosine_colors)]
        linestyle = line_styles["cosine"]
    else:
        color = "gray"
        linestyle = "-"
    inset_ax.plot(epochs, scores, marker="o", label=model, color=color, linestyle=linestyle)

inset_ax.set_xlim(50, 75)  # Zoom in on epochs 50 to 75
inset_ax.set_ylim(10, 20)  # Adjust Y-axis to focus on the zoomed data range
inset_ax.set_title("Zoomed: Epochs 50-75", fontsize=10)
inset_ax.grid()

# Highlight inset region
ax.indicate_inset_zoom(inset_ax, edgecolor="black", linewidth=1.5)

plt.show()
```
