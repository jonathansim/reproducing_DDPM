## Confidence intervals

```python
import numpy as np
from scipy.stats import t

# Example data: Replace with your actual metrics for each experiment
experiments = {
    #"L1": [100, 100, 100],
    "L4": [12.62, 11.20, 8.08],
    "L8": [9.081, 10.69, 6.99],
    #"C1": [23.8, 24.1, 23.6],
    "C4": [6.71, 8.22, 6.08],
    #"C8": [23.8, 24.1, 23.6],
    # Add all experiments
}

# Confidence interval computation
ci_results = {}
confidence_level = 0.90
margin_of_errors = []

for exp, values in experiments.items():
    mean = np.mean(values)
    std_dev = np.std(values, ddof=1)  # Sample standard deviation
    sem = std_dev / np.sqrt(len(values))
    t_critical = t.ppf((1 + confidence_level) / 2, df=len(values) - 1)
    margin_of_error = t_critical * sem
    margin_of_errors.append(margin_of_error)
    ci = (mean - margin_of_error, mean + margin_of_error)
    ci_results[exp] = {"Mean": mean, "CI": ci}

# Print results
for i, (exp, stats) in enumerate(ci_results.items()):
    print(f"{exp}: Mean = {stats['Mean']:.2f}, {confidence_level*100}% CI = ({stats['CI'][0]:.2f}, {stats['CI'][1]:.2f}), MoE = {margin_of_errors[i]:.2f}")
```
